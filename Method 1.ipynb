# Light Estimation

$$
\min_{l} \sum_{(x,y) \in \Omega} \left( I(x,y) - \rho_{ref}(x,y) l^T \hat{n}_{ref}(x,y) \right)^2
$$


 where $\hat{n} = (1, n_x, n_y, n_z)^T$

# In vector form


$$
|| \vec{i} - \vec{\rho}_{ref}\circ  N_{ref}^T \cdot \vec{l} || ^2
$$




```python
as_vec = lambda x : np.array([x.flatten()]).T
```

```python
def rearrange_normals(N,img,albedo):
    I = as_vec(img)
    Pref = as_vec(albedo)
    (m,n,_) =  N.shape
    nx = np.zeros((m,n))
    ny = np.zeros((m,n))
    nz = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            nx[i][j],ny[i][j],nz[i][j] =  ps_n[i][j][:]
    [nx,ny,nz] = [as_vec(v).T for v in [nx,ny,nz]]
    ones = np.ones_like(nx)
    return np.concatenate([ones,nx,ny,nz])
```

```python
def estimate_light(img,N,albedo):
    Nref = rearrange_normals(N,img,albedo)
    Pref = albedo.ravel()
    I = img.ravel()
    c0 = 1/np.sqrt(4*np.pi)
    c1 = np.sqrt(float(3)/float(4*np.pi))
    c = np.array([c0,c1,c1,c1])
    a0 = np.pi
    a1 = 2*np.pi/np.sqrt(3)
    a = np.array([a0,a1,a1,a1])
    nprime = np.linalg.pinv(Pref.ravel()*Nref)
    l = np.dot(I.T, nprime).ravel()/c
    return l
    
def normalize_harmonics(light):
    light += light[0]
    w = light[-3:] / np.linalg.norm(light[-3:])
    return normalise_vector(w[-3:])

def normalise_vector(v):
    return v / np.sqrt(np.sum(v ** 2, axis=-1))[..., None]
```

```python
import numpy as np
def synthetic_sphere(light):
    import warnings
    warnings.filterwarnings('ignore')
    # sphere radius
    r = 50
    # sphere grid 
    [x,y] = np.mgrid[-1.3*r*10:1.3*r*10,-1.3*r*10:1.3*r*10]
    # using finner grid to enchance resolution
    [x,y] = [0.1*x.astype(float),0.1*y.astype(float)]
    # albedo
    albedo = 0.5
    # illumination direction
    l = light
    # partial derivatives in x and y direction
    p = -x/np.sqrt(r*r-(x*x+y*y))
    q = -y/np.sqrt(r*r-(x*x+y*y))
    # calculate brightness
    E = albedo/np.sqrt(1+p*p + q*q) * (-l[0]*p - l[1]*q + l[2])
    # remove points not in the sphere
    E *= (r*r - (x*x + y*y)) >= 0
    # set negarive and nan brightness to 0
    E[E<0] = 0
    E[np.isnan(E)] = 0
    # normalize
    E /= E.max()
    return E

def plot_light(light):
    E = synthetic_sphere(light)
    %matplotlib inline  
    import matplotlib.cm as cm
    import matplotlib.pyplot as plt

    plt.imshow(E,cmap = cm.Greys_r)
    plt.axis('off')
    plt.show()
```
